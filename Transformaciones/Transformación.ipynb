```python
import pandas as pd

df1 = pd.read_csv('netflix_titles-score.csv')   ######## Create a DataFrame with the CSV.
```

```python
#### Change empty values to "G" 

df1['rating'] = df1['rating'].fillna('G')
```

```python
### Split the "duration" column into two.

df1[['duration_int','duration_type']] = df1.duration.str.split(expand = True)
```

```python
### added id column

df1.insert(0,"id","n"+df1.show_id)
```

```python
### Convert all strings to lowercase.

df1['type'] = df1['type'].str.lower()
df1['title'] = df1['title'].str.lower()
df1['director'] = df1['director'].str.lower()
df1['cast'] = df1['cast'].str.lower()
df1['country'] = df1['country'].str.lower()
df1['rating'] = df1['rating'].str.lower()
df1['listed_in'] = df1['listed_in'].str.lower()
df1['description'] = df1['description'].str.lower()
df1['duration_type'] = df1['duration_type'].str.lower()
```

```python
### modify dates format

df1['date_added'] = pd.to_datetime(df1['date_added'])
```

```python
### Normalize Seasons

df1['duration_type'] = df1['duration_type'].replace("seasons","season")
```

```python
### store the csv
 
df1.to_csv('netflix.csv')
```
